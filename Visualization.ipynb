```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv(r'C:\Users\pc\Prashanth\ToyotaCorolla.csv')
x=data['KM']
y=data['Price']
z=data['Weight']
plt.scatter(x,y,c=z,cmap='summer',edgecolor='black',linewidth=1,alpha=0.75)
cbar=plt.colorbar()
cbar.set_label("KM/Price")
plt.title("Price of Toyota Corolla")
plt.xlabel('KM')
plt.ylabel('Price')
plt.tight_layout()
plt.show()
```

```python
plt.boxplot([data["Price"],data["HP"],data["KM"]])
plt.xticks([1,2,3],["Price","HP","KM"])
plt.show()
```

```python
import seaborn as sns
sns.heatmap(data[["Price","KM","Doors","Weight"]].corr(),cmap="jet")
```

```python
x = data["KM"]
y = data["Weight"]
z = data["Price"]
plt.tricontour(x,y,z,levels=20,cmap="jet")
plt.colorbar(label='Price')
plt.title("Contour Plot")
plt.xlabel("KM")
plt.ylabel("Weight")
```

```python
from mpl_toolkits import mplot3d
x = data["KM"]
y = data["Doors"]
z = data["Price"]
ax = plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap="jet")
ax.set_title("3D Surface plot")
```
